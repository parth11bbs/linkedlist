```python
# Traversal in linked list
class Node:
    def __init__(self, data):
        self.data = data # n1.sata=5
        self.next = None # n1.next=None
        
class sll:
    def __init__(self):
        self.head = None # sll.head=None i.e. list is empty (we need to assign sll.head to n1,
                         # so that now head is pointing to n1)
        
    def traversal(self):
        if self.head is None:
            print("Singly list is empty")
        # When self.head is pointing to first node then only we can perform traversal
        else:
            a = self.head # Using a temp. variable so that head variable need not to be change
            while a is not None:
                print(a.data, end = " ")
                a = a.next
                
    def insert_at_beginning(self, data):
        print() # For changing the line
        nb = Node(data)
        nb.next = self.head # This will create a link between nb and n1 node
        self.head = nb # Changing head from n1 to nb
    
    
    def insert_at_end(self, data):
        print()
        ne = Node(data)
        a = self.head # For traversing
        while a.next is not None:
            a = a.next
        a.next = ne
    
    
    def insert_at_specified_node(self, position, data):
        print()
        nib = Node(data)
        a = self.head
        for i in range (1, position): # Position=1 is considered after the first node
            a = a.next
        nib.next = a.next
        a.next = nib
    
    
    def deletion_at_beginning(self):
        print()
        a = self.head
        self.head = a.next
        a.next = None
        
        
    def deletion_at_end(self):
        print()
        prev = self.head
        a = self.head.next
        while a.next is not None:
            a = a.next
            prev = prev.next
        prev.next = None
        
        
    def deletion_at_specified_node(self, position):
        print()
        prev = self.head
        a = self.head.next
        for i in range(1, position): # position=1 starts after first node
            a = a.next
            prev = prev.next
        prev.next = a.next
        a.next = None
                
                   
sll = sll()
n1 = Node(5) # Creating of n1 node
sll.head = n1 # Now head is assigned to n1 instead of None
n2 = Node(10) # Creating of n2 node (n1.data=10, n1.next=None)
n1.next = n2 # Connecting n1 & n2 node
n3 = Node(15)
n2.next = n3
n4 = Node(20)
n3.next = n4 # n4.next is automatically None, while calling Node n4
# Calling traversal function of sll class
sll.traversal() # sll is passed as a self
sll.insert_at_beginning(2)
sll.traversal()

sll.insert_at_end(25) # Inplace of self, sll is passed
sll.traversal()

sll.insert_at_specified_node(4, 7) 
sll.traversal()

sll.deletion_at_beginning()
sll.traversal()

sll.deletion_at_end()
sll.traversal()

sll.deletion_at_specified_node(2)
sll.traversal()
```

```python
# Doubly linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
        
class dll:
    def __init__(self):
        self.head = None
        
        
    def forward_traversal(self):
        if self.head is None:
            print("dll is empty")
        else:
            a = self.head
            while a is not None:
                print(a.data, end = " ")
                a = a.next
                
    def backward_traversal(self):
        print()
        if self.head is None:
            print("dll is empty")
        else:
            a = self.head
            while a.next is not None:
                a = a.next
            while a is not None:
                print(a.data, end = " ")
                a = a.prev
                
                
    def insertion_at_beginning(self, data):
        print()
        ns = Node(data)
        a = self.head
        a.prev = ns
        ns.next = a
        self.head = ns
        
    
    def insertion_at_end(self, data):
        ne = Node(data)
        a = self.head
        while a.next is not None:
            a = a.next
        a.next = ne
        ne.prev = a
        
    def insertion_at_specified_node(self, position, data):
        print()
        nib = Node(data)
        a = self.head
        for i in range(1, position):
            a = a.next
        nib.prev = a
        nib.next = a.next
        a.next.prev = nib
        a.next = nib
        
        
    def deletion_at_beginning(self):
        print()
        a = self.head
        self.head = a.next
        self.head.prev = None
        a.next = None
        
        
    def deletion_at_end(self):
        print()
        a = self.head.next
        before = self.head
        while a.next is not None:
            a = a.next
            before = before.next
        a.prev = None
        before.next = None
        
        
    def deletion_at_specified_position(self, position):
        print()
        a = self.head.next
        before = self.head
        for i in range(1, position):
            a = a.next
            before = before.next
        before.next = a.next
        a.next.prev = before
        a.next = None
        a.prev = None
                
                
n1 = Node(5)
dll = dll()
dll.head = n1
n2 = Node(10)
n2.prev = n1
n1.next = n2
n3 = Node(15)
n3.prev = n2
n2.next = n3
n4 = Node(20)
n4.prev = n3
n3.next = n4
dll.forward_traversal()
dll.backward_traversal()

dll.insertion_at_beginning(0)
dll.forward_traversal()

dll.insertion_at_end(25)
dll.forward_traversal()

dll.insertion_at_specified_node(4, 30)
dll.forward_traversal()

dll.deletion_at_beginning()
dll.forward_traversal()

dll.deletion_at_end()
dll.forward_traversal()

dll.deletion_at_specified_position(2)
dll.forward_traversal()
```
